```python
import pandas as pd
import os
from tqdm import tqdm
import geopandas as gpd
```

```python
output_path = r"D:\qing_research\1_flood_water\1_flood_water_network\data\repair\by_dma"
repair_path = r"D:\qing_research\1_flood_water\1_flood_water_network\data\repair\Repairs.xlsx"
DMA_mapping_path = r"D:\qing_research\1_flood_water\1_flood_water_network\data\Anglian_water\processed\DMA_Mapping.csv"
flood_2020_path = r"D:\qing_research\1_flood_water\1_flood_water_network\data\Geo\flood\2_extract_flood_date_waterflow\2020\flood_shp_2020.shp"
flood_2023_path = r"D:\qing_research\1_flood_water\1_flood_water_network\data\Geo\flood\2_extract_flood_date_waterflow\2023\flood_shp_2023.shp"
flood_2024_path = r"D:\qing_research\1_flood_water\1_flood_water_network\data\Geo\flood\2_extract_flood_date_waterflow\2024\flood_shp_2024.shp"
```

```python
DMA_mapping = pd.read_csv(DMA_mapping_path)
repair_data = pd.read_excel(repair_path)
flood_2020 = gpd.read_file(flood_2020_path)
flood_2023 = gpd.read_file(flood_2023_path)
flood_2024 = gpd.read_file(flood_2024_path)
```

```python
repair_data.drop(columns = ['Order NO','Unnamed: 2','Unnamed: 18','Unnamed: 19', 'Unnamed: 20','Day','Week', 'Month', 'Year'],inplace = True)
repair_data['repair_id'] = range(1,len(repair_data)+1)
repair_data = repair_data.merge(DMA_mapping, left_on='DMA', right_on='DMA_Name', how='left')
repair_data.drop(columns = ['DMA_Name','x_coord','y_coord'],inplace = True)
```

```python
repair_data['Basic Start Date'] = pd.to_datetime(repair_data['Basic Start Date'])
```

```python
floods_date_2020 = flood_2020.groupby(['start_date', 'end_date']).size().reset_index(name='counts_date')
```

```python
floods_date_2020['start_date'] = pd.to_datetime(floods_date_2020['start_date'])
floods_date_2020['end_date'] = pd.to_datetime(floods_date_2020['end_date'])
floods_date_2020['expanded_startdate'] = floods_date_2020['start_date'] - pd.Timedelta(days = 7)
floods_date_2020['expanded_enddate'] = floods_date_2020['end_date'] + pd.Timedelta(days = 14)
```

```python
flood_with_repair = []
for _,date_row in floods_date_2020.iterrows():
    start = date_row['expanded_startdate']
    end = date_row['expanded_enddate']
    flood_repair = repair_data[(repair_data['Basic Start Date']>=start)&(repair_data['Basic Start Date']<=end)]
    flood_with_repair.append(flood_repair)
```

```python
all_repair_flood_2020 = pd.concat(flood_with_repair, ignore_index=True)
all_repair_flood_2020 = all_repair_flood_2020.drop_duplicates()
print(len(all_repair_flood_2020))
```
