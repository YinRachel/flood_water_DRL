## funtion

```python
def generate_timerange(file_path):
    data = pd.read_csv(file_path,parse_dates=['time'])
    data['time'] = pd.to_datetime(data['time'])
    start_time = data['time'].min()
    year, month = start_time.year, start_time.month
    start_timerange = pd.Timestamp(year = year, month =month, day = 1, hour =0, minute =0 )
    end_timerange = (start_timerange + pd.offsets.MonthEnd(1)).replace(hour=23, minute=0)
    time_range = pd.date_range(start=start_timerange, end=end_timerange, freq="h")
    df_time = pd.DataFrame({"time": time_range})
    return df_time
```

```python
def read_data(base_path,datasets,month_num,example_path):
    for i in range(6):
        file_path = f"{base_path}/data_{i}.csv"
        data = pd.read_csv(file_path)      
        if i == 4:
            data_example = pd.read_csv(example_path)
            data['time'] = pd.to_datetime(data['time'])
            data = data[data['time'].dt.month == month_num].reset_index(drop=True)
            data['time'] = data_example['time']
        datasets.append(data)
        print(data.shape)
    return datasets
```

```python
def standardize_columns(datasets):
    for i, dataset in enumerate(datasets):
        dataset['latitude'] = dataset['latitude'].astype(float).round(2)
        dataset['longitude'] = dataset['longitude'].astype(float).round(2)
        dataset['time'] = pd.to_datetime(dataset['time'])
    reference = datasets[0]
    
    time_is_same = True
    latitude_is_same = True
    longitude_is_same = True
    for i in range(1, len(datasets)):
        time_is_same &= reference['time'].equals(datasets[i]['time'])
        latitude_is_same &= reference['latitude'].equals(datasets[i]['latitude'])
        longitude_is_same &= reference['longitude'].equals(datasets[i]['longitude'])
    print(f"All datasets have identical 'time' columns: {time_is_same}")
    print(f"All datasets have identical 'latitude' columns: {latitude_is_same}")
    print(f"All datasets have identical 'longitude' columns: {longitude_is_same}")
```

```python
def merge_data(datasets, on_columns):
    merged_data = datasets[0]
    for i, dataset in enumerate(datasets):
        datasets[i] = dataset.drop(columns=['number','step','depthBelowLandLayer','valid_time','surface'], errors='ignore')
    for i in range(1,len(datasets)):
        merged_data = merged_data.merge(datasets[i], on =on_columns,how='outer')
    merged_data = merged_data.drop(columns=['number','step','depthBelowLandLayer','valid_time'], errors='ignore')
    merged_data.rename(columns = columns_rename, inplace = True)
    return merged_data
```
