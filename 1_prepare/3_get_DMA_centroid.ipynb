```python
grid = np.arange(33 * 31).reshape(33, 31)
print("First row of the grid:", grid[0])
```

```python
def find_cell(lat, lon, lat_bnds, lon_bnds, grid):
    # Find row index (longitude) for the grid cell
    row_idx = np.searchsorted(lon_bnds, lon, side="right") - 1
    # Find column index (latitude) for the grid cell
    col_idx = np.searchsorted(lat_bnds, lat, side="right") - 1
    # Retrieve the cell number from the grid
    cell_number = grid[row_idx, col_idx]
    return cell_number
```

```python
data_centroid['cell_num'] = data_centroid.apply(
    lambda row: find_cell(row['latitude'], row['longitude'], lat_bnds, lon_bnds, grid),
    axis=1
)
```

```python
pattern = re.compile(r"(\d{8}-S\d{6})")

# Loop through each file in the folder
for filename in tqdm(os.listdir(precipitation_output_path),desc = "processing"):
    # Check if the filename matches the expected pattern
    match = pattern.match(filename)
    if match:
        # Extract the date-time part
        new_name = match.group(1)+ ".ascii"  # The part like "20200101-S000000"
        
        # Get the full path for the original and new file names
        original_file = os.path.join(precipitation_output_path, filename)
        new_file = os.path.join(precipitation_output_path, new_name)
        
        # Rename the file
        os.rename(original_file, new_file)
    else:
        print(f"Filename '{filename}' did not match the expected pattern.")
```

```python
from datetime import datetime, timedelta
```

```python
# Set the start and end dates
start_date = datetime(2020, 1, 1, 0, 0)
end_date = datetime(2020, 3, 31, 23, 30)

# Generate all expected timestamps in the format YYYYMMDD-SHHMMSS
expected_timestamps = []
current_time = start_date
while current_time <= end_date:
    # Format the timestamp into the expected filename format
    timestamp_str = current_time.strftime("%Y%m%d-S%H%M%S")
    expected_timestamps.append(timestamp_str)
    current_time += timedelta(minutes=30)

# Load actual filenames from the specified directory
actual_filenames = []
for filename in os.listdir(precipitation_output_path):
    # Check if the file has the .ascii extension
    if filename.endswith(".ascii"):
        # Split the filename by '.'
        parts = filename.split('.')[0]
        actual_filenames.append(parts)

# Find the missing filename(s)
missing_files = set(expected_timestamps) - set(actual_filenames)

# Print the missing datetime(s)
for missing_file in sorted(missing_files):
    print("Missing file:", missing_file)
```

## precipitation data to cell number

```python
all_cells_data = []
```

```python
def extract_precipitation(file_path):
    with open(file_path,'r') as file:
        content = file.read()
    precipitation_data = re.findall(r'precipitation\[\d+\]\[\d+\],\s*([\d.,\s]+)', content)
    grid = []
    for row in precipitation_data:
        values = [float(val) for val in row.split(',')]
        grid.append(values)
    
    return np.array(grid)
```

```python
date_range = pd.date_range(start = '2020-01-01 00:00:00', end = '2020-03-31 23:30:00', freq = '30min')
```

```python
from tqdm import tqdm
```

```python
dma_cell_path = "D:\\qing_research\\1_rainfall_water\\1_rainfall_water_network\\data\\precipitation\\dma_cell_num"
```

```python
for idx, (cell_num, dma_short_name) in tqdm(enumerate(zip(data_centroid['cell_num'],data_centroid['DMASHORTNA'])), total = len(data_centroid),desc="Processing Cells"):
#for cell_num in tqdm(data_centroid['cell_num'].unique(),desc="Processing Cells") :
    cell_precipitation_data = []
    for filename in sorted(os.listdir(precipitation_output_path)):
        if filename.endswith('.ascii'):
            file_path = os.path.join(precipitation_output_path,filename)
            grid = extract_precipitation(file_path)
            row_idx = cell_num // 31
            col_idx = cell_num % 31
            cell_precipitation_data.append(grid[row_idx][col_idx])
    cell_df = pd.DataFrame({
        'DateTime':date_range,
        'precipitation':cell_precipitation_data,
        'cell_num': cell_num
    })
    dma_cell_df = pd.DataFrame({
        'DateTime':date_range,
        'precipitation':cell_precipitation_data
    })
    filename = f"{dma_cell_path}/{dma_short_name}_{cell_num}.csv"
    dma_cell_df.to_csv(filename, index =False)
    all_cells_data.append(cell_df)
```
